#### Missing Value Identification and Replacing

```python
import numpy as np

# replace "?" to NaN
df.replace("?", np.nan, inplace = True)
df.head(5)
```


**Replace by mean:**
* "normalized-losses": 41 missing data, replace them with mean</li>

* "bore": 4 missing data, replace them with mean</li>
* "horsepower": 2 missing data, replace them with mean</li>

**Replace by frequency:**

```python
avg_norm_loss = df["NORMALIZED_LOSSES"].astype("float").mean(axis=0)
df["NORMALIZED_LOSSES"].replace(np.nan, avg_norm_loss, inplace=True)
print("Average of normalized-losses:", avg_norm_loss)
```

```python
avg_bore=df['BORE'].astype('float').mean(axis=0)
df["BORE"].replace(np.nan, avg_bore, inplace=True)
print("Average of bore:", avg_bore)
```

```python
avg_stroke=df["STROKE"].astype('float').mean(axis=0)
df["STROKE"].replace(np.nan,avg_stroke,inplace=True)
print("Average stroke:", avg_stroke)
```

```python
avg_horsepower = df['HORSEPOWER'].astype('float').mean(axis=0)
df['HORSEPOWER'].replace(np.nan, avg_horsepower, inplace=True)
print("Average horsepower:", avg_horsepower)
```

```python
avg_peakrpm=df["PEAK_RPM"].astype('float').mean(axis=0)
df["PEAK_RPM"].replace(np.nan, avg_peakrpm, inplace=True)
print("Average peak rpm:", avg_peakrpm)
```

```python
df["NUM_OF_DOORS"].value_counts()
```

```python
most_freq=df['NUM_OF_DOORS'].value_counts().idxmax()
df["NUM_OF_DOORS"].replace(np.nan, most_freq, inplace=True)
```

```python
#drop row with blank values

df.dropna(subset=["PRICE"], axis=0, inplace=True)
df.reset_index(drop=True, inplace=True)
```
